```python
def parse(lines):
    res = {}
    for line in lines:
        name, action = line.split(': ')
        try:
            action = int(action)
        except ValueError:
            action = action.split(' ')
        res[name] = action
    return res

def resolve(name, monkeys):
    if not isinstance(monkeys[name], list):
        return monkeys[name]
    ops = {
        '+': op.add,
        '*': op.mul,
        '-': op.sub,
        '/': op.ifloordiv,
    }
    nameA, _op, nameB = monkeys[name]
    return ops[_op](resolve(nameA, monkeys), resolve(nameB, monkeys))
```

```python
def resolveB(name, monkeys):
    if name == 'humn':
        return -1
    if isinstance(monkeys[name], int):
        return monkeys[name]
    ops = {
        '+': op.add,
        '*': op.mul,
        '-': op.sub,
        '/': op.ifloordiv,
    }
    nameA, _op, nameB = monkeys[name]
    # print(nameA, _op, nameB)
    result_left = resolveB(nameA, monkeys)
    result_right = resolveB(nameB, monkeys)
    
    if (result_left) >= 0 and (result_right >=0):
        result = ops[_op](result_left, result_right)
        monkeys[name] = result
        return result
    
    if result_left >= 0:
        monkeys[name] = [result_left, _op, nameB]
    else:
        monkeys[name] = [nameA, _op, result_right]
    return -1

def create_tree(name, monkeys):
    if name == 'humn':
        return " ? "
    if isinstance(monkeys[name], int):
        return str(monkeys[name])
    ops = {
        '+': op.add,
        '*': op.mul,
        '-': op.sub,
        '/': op.ifloordiv,
    }
    nameA, _op, nameB = monkeys[name]
    # print(nameA, _op, nameB)
    # return ops, (create_tree(nameA, monkeys), create_tree(nameB, monkeys))
    return " ( " + create_tree(nameA, monkeys) + " " + _op + " " + create_tree(nameB, monkeys) + " ) "

def solve_for_x(name, result, monkeys):
    if name == 'humn':
        print('result: ', result)
        return result
    
    nameA, _op, nameB = monkeys[name]
    # print(nameA, _op, nameB)
    
    match _op:
        case '+':
            name, val = (nameA, nameB) if isinstance(nameB, int) else (nameB, nameA)
            return solve_for_x(name, result - val, monkeys)
        case '*':
            name, val = (nameA, nameB) if isinstance(nameB, int) else (nameB, nameA)
            return solve_for_x(name, result // val, monkeys)
        case '-':
            if isinstance(nameA, int):
                return solve_for_x(nameB, nameA - result, monkeys)
            return solve_for_x(nameA, nameB + result, monkeys)
        case '/':
            if isinstance(nameA, int):
                return solve_for_x(nameB, nameA // result, monkeys)
            return solve_for_x(nameA, nameB * result, monkeys)
        
```
