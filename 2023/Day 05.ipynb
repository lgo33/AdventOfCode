```python
def parse(data):
    seeds = []
    maps = []
    new_map = []
    map_names = []
    im = 0
    for line in data:
        if line.startswith('seeds:'):
            _, vals = line.split(": ")
            seeds = [int(val) for val in vals.split(' ') if val]
            continue
            
        if not line.strip():
            continue
            
        if 'map:' in line:
            if new_map:
                maps.append(sorted(new_map, key=lambda x: x[1]))
            map_names.append(line.split(' map:')[0])
            new_map = []
            continue
        
        # print(line)
        new_map.append([int(val) for val in line.split(' ')])
        
    maps.append(sorted(new_map, key=lambda x: x[1]))        
    return seeds, maps, map_names

def lookup_map(val, Map):
    for dst, src, rng in Map:
        if (val>= src) and (val < src+rng):
            return val - src + dst
    return val

def get_location(seed, maps):
    for m in maps:
        seed = lookup_map(seed, m)
    return seed
```

```python
# if there is an overlap between a seed range and a map create 2 split ranges
vals = sr.copy()
for m in mr:
    new_vals = []
    for vmin, vmax in vals:
        for mmin, mmax, delta in m:
            if mmax < vmin:
                continue
            if vmax < mmin:
                new_vals.append([vmin, vmax])
                break
            
            if vmin < mmin:
                new_vals.append([vmin, mmin])
                vmin = mmin
                            
            if vmax < mmax:
                new_vals.append([vmin+delta, vmax+delta])
                break
            else:
                new_vals.append([vmin+delta, mmax + delta])
                vmin = mmax
                continue
        else:
            new_vals.append([vmin, vmax])
    vals = sorted(new_vals)
    print(vals)
```
