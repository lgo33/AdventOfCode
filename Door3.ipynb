```python
f = open('day3.txt', 'r')
inp = f.read()
f.close()
x = 0
y = 0
def hash():
    return str(x) + "|" + str(y)
```

```python
visited = {}
visited[hash()] = True
for c in inp[:]:
    if   c == '^':
        y += 1
    elif c == 'v':
        y -= 1
    elif c == '<':
        x -= 1
    elif c == '>':
        x += 1
    visited[hash()] = True
print len(visited)        
```

```python
visited = {}
visited[hash()] = True
for route in (inp[::2], inp[1::2]):
    x = 0
    y = 0
    for c in route:
        if   c == '^':
            y += 1
        elif c == 'v':
            y -= 1
        elif c == '<':
            x -= 1
        elif c == '>':
            x += 1
        visited[hash()] = True
print len(visited) 
```
