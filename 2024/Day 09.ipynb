```python
from collections import deque, defaultdict
```

```python
def parse(inp):
    blocks = []
    is_file = True
    idx = 0
    for digit in inp:
        d = int(digit)
        if is_file:
            blocks += [idx] * d
            idx += 1
        else:
            # print("!")
            blocks += [-1] * d
        is_file = not is_file
    return deque(blocks)

def compress(blocks):
    idx = blocks.index(-1)
    while True:
        # print(blocks)
        # get last element != -1
        el = -1
        while el == -1:
            el = blocks.pop()

        blocks[idx] = el
        
        # set next index
        while blocks[idx] != -1:
            idx += 1
            if idx >= len(blocks) - 1:
                if blocks[-1] == -1:
                    blocks.pop()
                return blocks
            
def chk_sum(blocks):
    return sum([i*el for i,el in enumerate(blocks)])
```

```python
# blocks = parse(test_data)
blocks = parse(data[0])
blocks = compress(blocks)
res = chk_sum(blocks)
res
```

```python
puzzle.answer_a = res
```

```python
def parse(inp):
    blocks = []
    is_file = True
    idx = 0
    files = {}
    gaps = defaultdict(deque)
    for digit in inp:
        d = int(digit)
        if is_file:
            files[idx] = (len(blocks), d)
            blocks += [idx] * d            
            idx += 1
        else:
            # print("!")
            if d > 0:
                gaps[d].append(len(blocks))
                blocks += [-1] * d
        is_file = not is_file
    return deque(blocks), gaps, files


def compress(gaps, files):
    N = max(files.keys())
    while N >= 0:
        pos, fsize = files[N]
        
        min_pos = len(blocks)
        chosen_gap_size = None
        # find gap        
        for i in range(fsize, 10):
            if len(gaps[i]) == 0:
                continue
            if gaps[i][0] < min_pos and gaps[i][0] < pos:
                min_pos = gaps[i][0]
                chosen_gap_size = i
        
        # found a gap
        if chosen_gap_size is not None:
            gap_pos = gaps[chosen_gap_size].popleft()            
            files[N] = (gap_pos, fsize)
            if chosen_gap_size > fsize:
                # reinsert smaller gap at right position
                new_gap_pos = gap_pos + fsize
                new_gap_size = chosen_gap_size - fsize
                i = 0
                while gaps[new_gap_size][i] < new_gap_pos:
                    i += 1
                gaps[new_gap_size].insert(i, new_gap_pos)
        # print(gaps)
        N -= 1
    return files, gaps
def chk_sum(files):
    res = 0
    for idx, file in files.items():
        pos, size = file
        for i in range(size):
            res += (pos+i) * idx
    return res
        
```

```python
blocks, gaps, files = parse(data[0])
# blocks, gaps, files = parse(test_data)
files, gaps = compress(gaps, files)
res = chk_sum(files)
res
```
