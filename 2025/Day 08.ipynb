```python
# pos = np.array([list(map(int, dat.split(','))) for dat in data_test])
pos = np.array([list(map(int, dat.split(','))) for dat in data])

inf = int(1e12)

dist = ((pos - pos.reshape(-1, 1, 3))**2).sum(-1)
dist[dist == 0] = inf

G = nx.Graph()

for i in range(1000):
    id1, id2 = sorted(np.unravel_index(dist.argmin(), dist.shape))
    dist[id1, id2] = inf
    dist[id2, id1] = inf
    G.add_edge(id1, id2)
    # print(pos[id1], pos[id2])
    # print(id1, id2)
```

```python
ans = reduce(mul, sorted([len(component) for component in nx.connected_components(G)])[-3:])
ans

```

```python
puzzle.answer_a = ans
```
