```python
def parse(ins):
    step = int(ins[1:])
    if ins[0] == 'L':
        step *= -1
    return step
```

```python
dial = 50
N = 100
cnt = 0

for i in data:
    dial += parse(i)
    dial %= N
    if dial == 0:
        cnt +=1
print(cnt)
```

```python
# is this the gemini autocomplete? Answer: 42   
from functools import cache

@cache
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```
