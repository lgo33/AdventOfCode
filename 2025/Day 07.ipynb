```python
test_data = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
..............."""
data_test = test_data.splitlines()
```

```python
data = puzzle.input_data.splitlines()
len(data), data[:10]
```

```python
def parse(data):
    S = None
    splitters = set()
    for r, row in enumerate(data):
        for c, val in enumerate(row):
            if val == 'S':
                S = (c, r)
            elif val == '^':
                splitters.add((c, r))
    return S, splitters
```

```python
# dat = data_test
dat = data

start, splitters = parse(dat)
Y = len(dat)

stack = [start]
visited = set()
splits = 0
while stack:
    x, y = stack.pop()
    if (x, y) in visited:
        continue
    visited.add((x, y))
    # down
    if (x, y) in splitters:
        splits += 1
        for d in [-1, 1]:
            if (x + d, y) in visited:
                
                continue
            stack.append((x + d, y))
            # print(f"Split at {(x, y)} to {(x + d, y)}")
            
    elif y + 1 < Y:
        stack.append((x, y + 1))
        
splits
        
```

```python
puzzle.answer_a = str(splits)
```

```python
@cache
def npaths(x, y, Y, splitters):
    if y == Y:
        return 1
    if (x, y) not in splitters:
        return npaths(x, y + 1, Y, splitters)
    total = 0
    for d in [-1, 1]:
        total += npaths(x + d, y, Y, splitters)
    
    return total
```

```python
# dat = data_test
dat = data

start, splitters = parse(dat)
Y = len(dat)

tot = npaths(start[0], start[1], Y, frozenset(splitters))
print(tot)
```

```python
puzzle.answer_b = str(tot)
```
