```python
def parse(rng):
    low, high = rng.split('-')
    return int(low), int(high)

def generate_invalid(low, high):
    strlow = str(low)
    strhigh = str(high)
    N = len(strlow)
    if N%2:
        start = '1' + '0' * (N//2)
    else:
        start = strlow[:N//2]
    cnt = 0
    idx = int(start)
    while True:
        cand = int(str(idx) * 2)
        if cand < low:
            idx += 1
            continue
        if cand <= high:
            yield(cand)
            cnt += 1
            idx += 1
        else:
            break
```

```python
# dat = test_data.split(',')
dat = data[0].split(',')

tot = 0
for d in dat:
    low, high = parse(d)
    for c in generate_invalid(low, high):
        tot += c
print(tot)
```

```python
puzzle.answer_a = str(tot)
```

```python
def divisors(x):
    for i in range(2, x//2+1):
        if (x%i) == 0:
            yield i
    yield x
            
def split_n(string, n):
    N = len(string)
    j = N//n
    return [string[i*j: (i+1)*j] for i in range(n)]

def chk(arr):
    for i in range(len(arr)):
        if arr[-1] != arr[i]:
            return False
    return True
    

def generate_invalid_2(low, high):
    for i in range(low, high+1):
        si = str(i)
        N = len(si)
        if N < 2:
            continue
        for n in divisors(N):
            if chk(split_n(si, n)):
                yield i
```

```python
# dat = test_data.split(',')
dat = data[0].split(',')

tot = set()
for d in dat:
    low, high = parse(d)
    invalid = set(generate_invalid_2(low, high))
    print(low, high, invalid)
    tot.update(invalid)
```
