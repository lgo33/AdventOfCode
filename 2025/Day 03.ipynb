```python
def parse(line):
    return [int(x) for x in line]

def max_joltage(bat):
    i = max(bat[:-1])
    idx = bat.index(i)
    j = max(bat[idx+1:])
    return int(str(i) + str(j))
```

```python
# batteries = [parse(line) for line in data_test]
batteries = [parse(line) for line in data]

joltages = [max_joltage(bat) for bat in batteries]
tot = sum(joltages)
tot
```

```python
puzzle.answer_a = str(tot)
```

```python
def get_max_digits(bat, n):
    if n == 0:
        return [str(max(bat))]
    i = max(bat[:-n])
    idx = bat.index(i)
    return [str(i)] + get_max_digits(bat[idx+1:], n-1)
```

```python
arr = [1, 2, 2, 5, 5, 4]
arr[:-1]
```

```python
# batteries = [parse(line) for line in data_test]
batteries = [parse(line) for line in data]

tot = sum([int(''.join(get_max_digits(bat, 11))) for bat in batteries])
tot
```

```python
puzzle.answer_b = str(tot)
```
