```python
data = puzzle.input_data.splitlines()
len(data), data[:10]
```

```python
def parse(data):
    shapes = []
    shape = []
    parse_shapes = True
    regions = []
    for line in data:
        if parse_shapes:
            if line == '':
                shapes.append(shape)
                shape = []
                if len(shapes) >= 6:
                    parse_shapes = False
            elif line[-1] == ':':
                continue 
            shape.append([1 if c =='#' else 0 for c in line])
        else:
            if line[-1] == '':
                continue 
            size, num_shapes = line.split(': ')
            size = list(map(int, size.split('x')))
            num_shapes = list(map(int, num_shapes.split(' ')))
            regions.append((size, num_shapes))
    return shapes, regions
    
```

```python
shapes, regions = parse(data)
shape_sizes = [sum([sum(row) for row in shape]) for shape in shapes]
```

```python
shape_sizes
```

```python
tot = 0
for size, num_shapes in regions:
    total_area = size[0] * size[1]
    shape_area = 0
    for i, num in enumerate(num_shapes):
        shape_area += num * shape_sizes[i]
    if shape_area <= total_area:
        tot += 1
tot
```

```python
puzzle.answer_a = int(tot)
```
