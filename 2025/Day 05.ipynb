```python
def parse(lines):
    ranges = []
    ids = []
    parse_ranges = True
    for line in lines:
        if line == '':
            parse_ranges = False
            continue
        if parse_ranges:
            a,b = line.split('-')
            ranges.append( (int(a), int(b)) )
        else:
            ids.append( int(line) )
    return ranges, ids
```

```python
# ranges, ids = parse(data_test)
ranges, ids = parse(data)

tot_fresh = 0
tot_spoiled = 0
for id in ids:
    for a,b in ranges:
        if a <= id <= b:
            tot_fresh += 1
            break
    else:
        tot_spoiled += 1
        
print(tot_fresh, tot_spoiled)
```

```python
puzzle.answer_a = str(tot_fresh)
```

```python
def merge_ranges(ranges):
    ranges = sorted(ranges)
    merged = []
    cur_a, cur_b = ranges[0]
    for a,b in ranges[1:]:
        if a <= cur_b + 1:
            cur_b = max(cur_b, b)
        else:
            merged.append( (cur_a, cur_b) )
            cur_a, cur_b = a, b
    merged.append( (cur_a, cur_b) )
    return merged
```

```python
# ranges, ids = parse(data_test)
ranges, ids = parse(data)

merged_ranges = merge_ranges(ranges)
tot_fresh = sum([b - a + 1 for a,b in merged_ranges])
merged_ranges, tot_fresh
```

```python
puzzle.answer_b = str(tot_fresh)
```
