```python
def render(scr):
    for c in scr:
        print(chr(c), end="")
```

```python
def split(_in):
    scr = _in.copy()
    lines = []
    line = []
    while scr:
        c = scr.pop(0)
        if c != 10:
            line.append(c)
        elif line:
            lines.append(line)
            line = []
    return lines
```

```python
lines = split(scr)
_y = len(lines)
_x = len(lines[0])
_x, _y
```

```python
intersections = []
for i in range(1, _x-1):
    for j in range(1, _y-1):
        for x, y in ((i, j), (i-1, j), (i+1, j), (i, j-1), (i, j+1)):
            if lines[y][x] != 35:
                break
        else:
            intersections.append((i, j))
```

```python
# part 1
sum([x * y for x,y in intersections])
```
