```python
def nextline(line):
    nline = []
    l = '.'
    c = line[0]
    for i in range(len(line)):
        r = line[i+1] if i < len(line)-1 else '.'
        n = '.'
        if l == '^' and r == '.':
            n = '^'
        elif l == '.' and r == '^':
            n = '^'
        nline.append(n)
        l = c
        c = r
    return ''.join(nline)
```

```python
l = '.^^.^.^^^^'
s = 0
for i in range(10):
    print l, l.count('.')
    s += l.count('.')
    l = nextline(l)
print s
```
