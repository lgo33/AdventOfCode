```python
import aocd
from aocd.models import Puzzle
day = 5
year = 2020
puzzle = Puzzle(year=year, day=day)
# data = aocd.get_data(day=day, year=year)
with open('./data/input_{:02d}'.format(day), 'w') as fh:
    fh.write(puzzle.input_data)
```

```python
data = puzzle.input_data.splitlines()
# data = test_input.split('\n\n')
```

```python
def getID(col, row):
    return 8*col + row

def parse(line):
    fb = line[:7].replace('B', '1').replace('F', '0')
    lr = line[7:].replace('R', '1').replace('L', '0')
    col, row = int(fb, 2), int(lr, 2)
    
    return col, row, getID(col, row)
```

```python
parse('BFFFBBFRRR')
```

```python
import numpy as np

seats = np.array(list(map(parse, data)))
```

```python
puzzle.answer_a = seats.max(axis=0)[2]
```

```python
# Part B
```

```python
sorted_ids = np.sort(seats[:, 2])
my_id = sorted_ids[np.diff(sorted_ids, append=False) == 2][0] + 1
```

```python
puzzle.answer_b = my_id
```
